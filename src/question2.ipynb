```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0, 20)  # 20 samples

# Frequencies to compare
omega1 = np.pi / 4  # Low frequency
omega2 = np.pi / 2  # Mid frequency
omega3 = np.pi      # Max frequency (highest oscillation)

# Sinusoidal signals
x1 = np.sin(omega1 * n)
x2 = np.sin(omega2 * n)
x3 = np.sin(omega3 * n)

# Plotting
plt.figure(figsize=(10, 10))

plt.subplot(3, 1, 1)
plt.stem(n, x1, basefmt=" ")
plt.title(r'$\omega = \pi/4$ (Low Oscillation)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(n, x2, basefmt=" ")
plt.title(r'$\omega = \pi/2$ (Medium Oscillation)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(n, x3, basefmt=" ")
plt.title(r'$\omega = \pi$ (Highest Oscillation)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
