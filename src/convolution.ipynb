```python
import numpy as np
import matplotlib.pyplot as plt


def convolve(x,h):
  len_x=len(x)
  len_h=len(h)
  len_y=len_x+len_h-1
  y=np.zeros(len_y)
  for n in range(len_y):
    for k in range(len_h):
      if n-k>=0 and n-k<len_x:
        y[n]+=x[n-k]*h[k]
  return y

def correlation(x,h):
  len_x=len(x)
  len_h=len(h)
  len_y=len_x+len_h-1
  y=np.zeros(len_y)
  for n in range(len_y):
    for k in range(len_h):
      if n-k>=0 and n-k<len_h:
        y[n]+=x[k]*h[n-k]
  return y

x=np.array([1,3,-2,4])
h=np.array([2,3,-1,3])

y=convolve(x,h)
yc=correlation(x,x)

plt.figure(figsize=(10,10))


plt.subplot(3,2,1)
plt.title('Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.stem(np.arange(len(x)),x)
plt.subplot(3,2,2)
plt.title('Signal h[n]')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.stem(np.arange(len(h)),h)

plt.subplot(3,2,3)
plt.title('Convolution of two signals')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.stem(np.arange(len(y)),y)

plt.subplot(3,2,4)
plt.title('Correlation of two signals')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.stem(np.arange(len(yc)),yc)


plt.tight_layout()
```
