```python
import numpy as np
import matplotlib.pyplot as plt

T=0.02

def f(t):
  return 3*np.cos(200*np.pi*t)+5*np.sin(600*np.pi*t)+10*np.sin(1200*np.pi*t)

fs1=1200

t1=np.arange(0,T,1/fs1)


x1=f(t1)

plt.figure(figsize=(10, 6))


tc=np.linspace(0,T,100)
xc=f(tc)
plt.subplot(3, 2, 1)
plt.plot(tc, xc, 'k', label='fs=1000Hz')
plt.grid()

plt.subplot(3, 2, 2)
plt.stem(t1, x1, 'r', label='fs=200Hz')
plt.grid()


plt.tight_layout()
plt.show()
```
