```python
import numpy as np
import matplotlib.pyplot as plt

T=0.2

def f(t):
  return 3*np.sin(100*np.pi*t)

fs1=200
fs2=75

t1=np.arange(0,T,1/fs1)
t2=np.arange(0,T,1/fs2)
t3=np.arange(0,T,1/500)
x1=f(t1)
x2=f(t2)
x3=f(t3)
plt.figure(figsize=(10, 6))


tc=np.linspace(0,T,100)
xc=f(tc)
plt.subplot(3, 2, 1)
plt.plot(tc, xc, 'k', label='fs=1000Hz')
plt.grid()

plt.subplot(3, 2, 2)
plt.stem(t1, x1, 'r', label='fs=200Hz')
plt.grid()

plt.subplot(3, 2, 3)
plt.stem(t2, x2, 'b', label='fs=75Hz')
plt.grid()

plt.subplot(3, 2, 4)
plt.stem(t3, x3)
plt.grid()

plt.tight_layout()
plt.show()
```
