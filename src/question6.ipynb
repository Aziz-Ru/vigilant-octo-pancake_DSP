```python
import numpy as np
import matplotlib.pyplot as plt


x=np.array([1,4,-2,4])
y=np.array([2,3,-1,3])
z=np.array([2,-1,4,-2])



xy=np.convolve(x,y)
yz=np.convolve(y,z)

plt.figure(figsize=(10,10))

plt.subplot(4,2,1)
plt.stem(np.arange(0,len(x),1),x)
plt.title('Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid()

plt.subplot(4,2,2)
plt.stem(np.arange(0,len(y),1),y)
plt.title('Signal y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid()

plt.subplot(4,2,3)
plt.stem(np.arange(0,len(z),1),z)
plt.title('Signal z[n]')
plt.xlabel('n')
plt.ylabel('z[n]')
plt.grid()



plt.subplot(4,2,4)
plt.stem(np.arange(0,len(xy),1),xy)
plt.title('Convolution of x and y')
plt.xlabel('n')
plt.ylabel('xy[n]')
plt.grid()

plt.subplot(4,2,5)
plt.stem(np.arange(0,len(yz),1),yz)
plt.title('Convolution of y and z')
plt.xlabel('n')
plt.ylabel('yz[n]')
plt.grid()
plt.tight_layout()
plt.show()


```
