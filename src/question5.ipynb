```python
import numpy as np
import matplotlib.pyplot as plt

T=0.02

def unit_step(n):
    return np.where(n>=0,1,0)

n=np.arange(0,10,1)

x=unit_step(n)
h=unit_step(n-5)
y=np.convolve(x,h)

plt.figure(figsize=(10, 5))

plt.subplot(3, 1, 1)
plt.stem(n, x,)
plt.title('x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid()

plt.subplot(3, 1, 2)
plt.stem(n, h)
plt.title('h[n]')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.grid()

plt.subplot(3, 1, 3)
plt.stem(np.arange(0,len(y)),y)
plt.title('y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid()
```
