```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y(t)=x(t)+2*x(t-1)
t=np.linspace(0,100,500)
x=np.sin(t)
y=[0]*len(x)
for i in range(1,len(x)):
    y[i]=x[i]+2*x[i-1]


plt.figure(figsize=(8, 4))
plt.stem(t, x, linefmt='b-', markerfmt='bo', basefmt='k', label='Input x[n]')
plt.stem(t, y, linefmt='r-', markerfmt='ro', basefmt='k', label='Output y[n]')
plt.xlabel('Time index n')
plt.ylabel('Amplitude')
plt.title('Causal System: y[n] = x[n] + 2*x[n-1]')
plt.legend()
plt.grid(True)
plt.show()
```
