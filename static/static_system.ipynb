```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Define input signal
n=np.arange(0,11)
x=np.sin(0.5*n)    # Example input: Sine function
# print(n)
# Define static system y[n] = 3x[n] + 2
y = 3 * x + 2
```

```python
plt.figure(figsize=(10,4))
```

```python
# input signal
plt.subplot(2,2,(1,2))
plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('Input signal')
plt.grid()

plt.subplot(2,2,(3,4))
plt.stem(n,y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Output signal')
plt.grid()

plt.tight_layout()
plt.show()
```
