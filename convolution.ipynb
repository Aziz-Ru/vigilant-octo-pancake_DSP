```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(1,6)
# print(x)
y=np.array([1,2,-3,4])
```

```python
# perform convolution
z=np.convolve(x,y)

zz=np.correlate(x,y)
```

```python
# define time indeces
nx=np.arange(len(x))
ny=np.arange(len(y))
nz=np.arange(len(z))
nzz=np.arange(len(zz))
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(3,1,1)
plt.stem(nx,x)
plt.title('x[n]')

plt.subplot(3,2,1)
plt.stem(ny,y)
plt.title('y[n]')

plt.subplot(3,1,2)
plt.stem(nz,z)
plt.title('z[n]')

plt.subplot(3,1,3)
plt.stem(nzz,zz)
plt.title('zz[n]')

plt.tight_layout()

plt.show()
```
