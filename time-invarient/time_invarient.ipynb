```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.linspace(0, 10, 1000)
x_T = np.sin(T)
y_T = np.sin(x_T)
```

```python
plt.plot(T, y_T, label="y(T) = sin[x(T)]")
plt.xlabel("T")
plt.ylabel("y(T)")
plt.title("Plot of y(T) = sin[x(T)]")
plt.legend()
plt.grid()
plt.show()
```
