```python
import numpy as np
import matplotlib.pyplot as plt
# U(t)= 1 when t>=0 and U(t)=0 when t<0
```

```python

def unit_step(t):
    return 1.0 * (t >= 0)

T=np.arange(-5,5)
U_T=np.where(T>=3,1,0)

plt.plot(T,U_T)
plt.title('Unit Step Function')
plt.xlabel('t')
plt.ylabel('U(t)')
plt.grid()
plt.show()

```
