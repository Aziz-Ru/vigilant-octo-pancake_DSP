```python
import numpy as np
import matplotlib.pyplot as plt

# This signal tends to infinity when t = 0 and tends to zero when t ≠ 0 such that the area under its curve is always equals to one
```

```python
T=np.arange(-10,10,1)
impulse=np.zeros_like(T)
impulse[T==0]=1
# print(impulse)
# set the impulse at t=0 to 1

plt.stem(T,impulse)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Unit Impulse Signal')
plt.grid()
plt.show()
```
