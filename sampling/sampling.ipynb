```python
import numpy as np
import matplotlib.pyplot as plt


# Continous signal
T=np.linspace(0,1,1000)
freq=5
amplitude=10
continious_signal=amplitude*np.sin(2*np.pi*freq*T)

plt.figure(figsize=(20,5))

plt.subplot(2,1,1)
plt.plot(T,continious_signal)
plt.title('Continious Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()

# Sampling
fs=200
Ts=1/fs
sample_time=np.arange(0,1,Ts)
sampled_signal=amplitude*np.sin(2*np.pi*freq*sample_time)

plt.subplot(2,1,2)
plt.stem(sample_time,sampled_signal)
plt.title('Sampled Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()


```
