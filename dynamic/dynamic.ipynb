```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
# Define Newton's Law of Cooling
def cooling_law(t, T, k, T_ambient):
    return -k * (T - T_ambient)
    
```

```python
# Parameters
T_initial = 90  # Initial temperature of coffee (°C)
T_ambient = 25  # Room temperature (°C)
k = 0.1         # Cooling rate
```

```python
# Time range for simulation
t_values = np.linspace(0, 100, 300)  # 100 time units

# Solve the differential equation
sol = solve_ivp(lambda t, T: cooling_law(t, T, k, T_ambient), (0, 100), [T_initial], t_eval=t_values)

# Plot the cooling curve
plt.figure(figsize=(8, 5))
plt.plot(sol.t, sol.y[0], label="Coffee Temperature", color="b", linewidth=2)

# Add ambient temperature line
plt.axhline(T_ambient, color="r", linestyle="--", label="Room Temperature (25°C)")
```

```python
# Labels and title
plt.xlabel("Time (minutes)")
plt.ylabel("Temperature (°C)")
plt.title("Dynamic System: Cooling of Coffee")
# plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
