```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define parameters for the linear system (spring-mass system)
m = 1.0   # Mass (kg)
k = 1.0   # Spring constant (N/m)
omega = np.sqrt(k / m)  # Natural frequency

# Time range for simulation
t_values = np.linspace(0, 20, 300)  # Simulate for 20 seconds

# Initial conditions
x0 = 1.0  # Initial displacement (m)
v0 = 0.0  # Initial velocity (m/s)

# Solve the differential equation: x'' + (k/m)x = 0
sol = solve_ivp(lambda t, y: [y[1], -omega**2 * y[0]], (0, 20), [x0, v0], t_eval=t_values)

# Plot the oscillations
plt.figure(figsize=(8, 5))
plt.plot(sol.t, sol.y[0], label="Displacement (x)", color="b", linewidth=2)

# Labels and title
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.title("Linear System: Spring-Mass Oscillation")
plt.axhline(0, color="r", linestyle="--", label="Equilibrium Position")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
