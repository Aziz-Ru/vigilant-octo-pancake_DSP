```python
import numpy as np
import matplotlib.pyplot as plt

# Define time range
T = np.linspace(0, 10, 1000)  # Example: 0 to 10 with 1000 points

# Define x(T) - example function, you can modify this based on your requirements
x_T = np.sin(T)  # Example: x(T) = sin(T)

# Calculate y(T) = cos(T) * x(T)
y_T = np.cos(T) * x_T

# Plot the results
plt.plot(T, y_T, label=r"$y(T) = \cos(T) \cdot x(T)$")
plt.xlabel("T")
plt.ylabel("y(T)")
plt.title("Plot of y(T) = cos(T) * x(T)")
plt.legend()
plt.grid(True)
plt.show()
```
