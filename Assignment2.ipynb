```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolve(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = [0] * len_y
    for i in range(len_y):
        for k in range(len_x):
            if(i - k >= 0 and i - k < len_h):
                y[i] += x[k] * h[i - k]
    return y
```

```python
x = np.array([4, 1, 2, 5]) 
e = np.arange(0, 4)

h = np.array([1, 2, -1])
y = np.arange(-1, 2)

s = convolve(x, h) 
w = np.arange(-1, len(x) + len(h) - 2)

plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.stem(e, x)
plt.xlabel('Time index')
plt.ylabel('Amplitude')
plt.title('x(n) function')
plt.axis([-2, 5, 0, 6])

plt.subplot(3, 1, 2)
plt.stem(y, h)
plt.xlabel('Time index')
plt.ylabel('Amplitude')
plt.title('h(n) function')
plt.axis([-2, 5, -2, 3])

plt.subplot(3, 1, 3)
plt.stem(w, s)
plt.xlabel('Time index')
plt.ylabel('Amplitude')
plt.title('Discrete Convoluted Signal')
plt.axis([-2, 5, -7, 11])
plt.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```
